### 1.1. O Problema de Negócio

A _churn rate_ (taxa de rotatividade ou cancelamento) é a porcentagem de assinantes de um serviço que interrompe suas assinaturas dentro de um determinado período de tempo. Para uma empresa expandir sua clientela, sua taxa de crescimento, medida pelo número de novos clientes, deve exceder sua taxa de cancelamento (_churn rate_).

Portanto, neste projeto, um banco deseja que você construa um modelo de Machine Learning que possa prever a probabilidade de seus clientes atuais fecharem suas contas em um futuro próximo e, portanto, calcular sua _churn rate_.

Esse é um problema de classificação no qual você classifica se um cliente irá sair (y=1) ou permanecer (y=0) com base diversos atributos.

### 1.2. Descrição do dados

No arquivo *Churn_Modelling.csv* você encontrará:
- **RowNumber**: id do registro/linha.
- **CustomerId**: identificador único do cliente.
- **Surname**: sobrenome do cliente.
- **CreditScore**: pontuação de crédito do cliente.
- **Geography**: país de origem do cliente.
- **Gender**: sexo do cliente.
- **Age**: idade do cliente.
- **Tenure**: número de anos em que o cliente está com o banco.
- **Balance**: saldo bancário.
- **NumOfProducts**: número de produtos que o cliente utiliza.
- **HasCrCard**: indica se o cliente tem (1) ou não tem (0) cartão de crédito.
- **IsActiveMember**: indica se o cliente é (1) ou não é (0) ativo no banco.
- **EstimatedSalary**: salário estimado em dólares.
- **Exited**: indica se o cliente encerrou (1) a conta no banco ou se o cliente foi retido (0).

### 3.1 Análise exploratória

```python
# seu código
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
```

### 3.2 Preparação e limpeza dos dados

```python
# seu código
df['Geography'].value_counts()
### Só 3 países
dft = df
```

```python
# seu código
dft = pd.get_dummies(df, columns=['Gender','Geography'])
```

```python
dft.info()
```

### 3.3 Seleção de atributos

```python
# seu código
X = dft.drop('Exited',axis=1)
y = dft.Exited.T

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.4, random_state=42)
```

```python
# seu código
```

### 3.4 Construção do modelo

```python
#Logistic Regression Classification
#K-Nearest Neighbour (KNN) Classification
#Support Vector Machine (SVM) Classification
#Naive Bayes Classification
#Decision Tree Classification
#Random Forest Classification

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier


```

```python
# LogReg

logreg = LogisticRegression()
logreg.fit(X_train, y_train)
logreg.score(X_test, y_test)
```

```python
logreg.classes_  
```

```python
#K-Nearest Neighbour (KNN) Classification
#knn3 = KNeighborsClassifier(1)
#knn3.fit(X_train, y_train)
#knn3.score(X_test, y_test)

score = []
for i in range(1,30):
    a = KNeighborsClassifier(i).fit(X_train, y_train)
    a.fit(X_train, y_train)
    score.append(a.score(X_test, y_test))

```

```python
plt.plot(range(1,len(score)+1),score)
plt.axis([1,20,0.5,1])
plt.xlabel("parâmetro N")
plt.ylabel("Acurácia")
plt.title("KNN - N x Acurácia")
```

```python
resultados = {}
resultados['LogReg'] = logreg.score(X_test, y_test)
resultados['KNN'] = max(score)

print(resultados)
```

```python
#Support Vector Machine (SVM) Classification

svc = SVC(gamma='scale')
svc.fit(X_train, y_train)
svc.score(X_test, y_test)

resultados['SVM'] = svc.score(X_test, y_test)
print(resultados)
```

```python
#Naive Bayes Classification
from sklearn.naive_bayes import GaussianNB

nbc = GaussianNB()
nbc.fit(X_train, y_train)
resultados['Naive Bayes Gauss'] = nbc.score(X_test, y_test)

print(resultados)
```

```python
#Decision Tree Classification
dtc =  DecisionTreeClassifier(max_depth=5)
dtc.fit(X_train, y_train)
dtc.score(X_test, y_test)

DecisionTreeClassifier()

scoredtc = []
for i in range(1,30):
    a = DecisionTreeClassifier(max_depth=i).fit(X_train, y_train)
    a.fit(X_train, y_train)
    scoredtc.append(a.score(X_test, y_test))

plt.plot(range(1,len(scoredtc)+1),scoredtc)
plt.axis([1,20,0.5,1])
plt.xlabel("parâmetro max depth")
plt.ylabel("Acurácia")
plt.title("Decision Tree Class - max depth x Acurácia")

resultados['Decision Tree Class max_depth=5'] = max(scoredtc)
```

```python
#Random Forest Classification

rfc = RandomForestClassifier(max_depth=5, n_estimators=100, max_features='auto')
rfc.fit(X_train, y_train)
rfc.score(X_test, y_test)


scorerfc2 = []
for i in range(1,10):
    a = RandomForestClassifier(max_depth=i, n_estimators=100, max_features='auto').fit(X_train, y_train)
    a.fit(X_train, y_train)
    scorerfc2.append(a.score(X_test, y_test))

plt.plot(range(1,len(scorerfc2)+1),scorerfc2)
plt.axis([1,10,0.5,1])
plt.xlabel("parâmetro max depth")
plt.ylabel("Acurácia")
plt.title("Random Forest Classification - max depth x Acurácia")
```

```python
resultados['Random Forest Classification max_depth=5'] = scorerfc2[4]
```
